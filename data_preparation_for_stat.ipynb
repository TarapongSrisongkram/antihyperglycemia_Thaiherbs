```python
import numpy as np
import pandas as pd
import scipy.stats
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx
```

```python
df= pd.read_csv('df_bioactivities.csv', index_col='Names')
df
```

```python
indices = df.index
```

```python
Emax_amylase = ['Emax_amylase_1', 'Emax_amylase_2','Emax_amylase_3']
df_Emax_amylase = pd.DataFrame(df[Emax_amylase], index=indices)
#convert matrix to list of edges and rename the columns
dft_Emax_amylase = df_Emax_amylase.stack().reset_index()
dft_Emax_amylase.columns = ['Names','Bioactivities','Emax_amylase']
df_Emax_amylase_for_stat = dft_Emax_amylase.drop('Bioactivities', axis=1)
df_Emax_amylase_for_stat = df_Emax_amylase_for_stat.set_index('Names')
df_Emax_amylase_for_stat
```

```python
Emax_glucosidase = ['Emax_glucosidase_1', 'Emax_glucosidase_2','Emax_glucosidase_3']
df_Emax_glucosidase = pd.DataFrame(df[Emax_glucosidase], index=indices)
dft = df_Emax_glucosidase.stack().reset_index()
dft.columns = ['Names','Bioactivities','Emax_glucosidase']
df_Emax_glucosidase_for_stat = dft.drop('Bioactivities', axis=1)
df_Emax_glucosidase_for_stat = df_Emax_glucosidase_for_stat.set_index('Names')
df_Emax_glucosidase_for_stat
```

```python
IC50_amylase = ['IC50_amylase_1', 'IC50_amylase_2','IC50_amylase_3']
df_IC50_amylase = pd.DataFrame(df[IC50_amylase], index=indices)
#convert matrix to list of edges and rename the columns
dft_IC50_amylase = df_IC50_amylase.stack().reset_index()
dft_IC50_amylase.columns = ['Names','Bioactivities','IC50_amylase']
df_IC50_amylase_for_stat = dft_IC50_amylase.drop('Bioactivities', axis=1)
df_IC50_amylase_for_stat = df_IC50_amylase_for_stat.set_index('Names')
df_IC50_amylase_for_stat.to_csv('IC50_amylase.csv')
df_IC50_amylase_for_stat
```

```python
IC50_glucosidase = ['IC50_glucosidase_1', 'IC50_glucosidase_2','IC50_glucosidase_3']
df_IC50_glucosidase = pd.DataFrame(df[IC50_glucosidase], index=indices)
#convert matrix to list of edges and rename the columns
dft_IC50_glucosidase = df_IC50_glucosidase.stack().reset_index()
dft_IC50_glucosidase.columns = ['Names','Bioactivities','IC50_glucosidase']
df_IC50_glucosidase_for_stat = dft_IC50_glucosidase.drop('Bioactivities', axis=1)
df_IC50_glucosidase_for_stat = df_IC50_glucosidase_for_stat.set_index('Names')
df_IC50_glucosidase_for_stat.to_csv('IC50_glucosidase.csv')
df_IC50_glucosidase_for_stat
```

```python
df_Emax = pd.concat([df_Emax_amylase_for_stat, df_Emax_glucosidase_for_stat], axis=1)
```

```python
df_Emax.to_csv('Emax_all.csv')
df_Emax
```
